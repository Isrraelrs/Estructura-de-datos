```python
numeroSecreto = 1997

print(
"""
+==================================+
| Bienvenido a mi juego, criatura! |
| Introduce un número entero       |
| y adivina qué número he          |
| elegido para ti.                 |
| Entonces,                        |
| ¿Cuál es el número secreto?      |
+==================================+
""")

numeroIngresado = int(input())

if numeroIngresado == numeroSecreto:
    print(numeroSecreto)
    print("¡Bien hecho, criatura! Eres libre ahora")
elif numeroIngresado != numeroSecreto:
    while True:
        print("¡que no puedes?! ¡Jamas podras salir!")
```
